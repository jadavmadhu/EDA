**Bar chart using seaborn**

- we can plot bar chart using many packages
- one package is matplotlib
- another one is seaborn
- in matplotlib the method is **plt.bar plot**
- in seaborn the method is **countplot**

- in matplotlib
  - we take the row data
  - we will create frequency table i.e DataFrame
  - we will pass the columns1, columns2, and DataFarme

- in seaborn
  - it takes direct raw data
  - we will provide direct columns name
  - we will provide main dataframe

```python
#continent colums ===== continent_df=======plt.bar
# continent column directly======visa_df


import seaborn as sns
sns.countplot(data=visa_df,x="continent")
```

```python
plt.figure(figsize=(10,5))
order_continent=['Asia','Europe','North America','South America','Africa','Oceania']
sns.countplot(data=visa_df,x="continent",order=order_continent)
plt.title("bar chart")
plt.savefig("continents_bar_chart_using_seaborn.jpg")
plt.show()
```

# PIE CHART

- pie chart is a 360 degree view
- it is under matplotlib package
- it will takes keys and values
- that we will get from value counts
  

```python
# step-1: get keys from continent value counts 
# step-2: get values from continent value counts
# step-3: plt.pie verify the argument
#            which argument takes values and keys
```

```python
keys=visa_df["continent"].value_counts().keys()
keys
```

```python
vals=visa_df["continent"].value_counts().values
vals
```

```python
keys=visa_df["continent"].value_counts().keys()
vals=visa_df["continent"].value_counts().values
plt.pie(x=vals,                              # values
        explode=[0.5,0.5,0.5,0.5,0.5,0.5],   # divides the pie
        labels=keys,                         # keys
        radius=2,                            # sixe of pie
        startangle=45,                       # angle to rotate
        autopct="%0.2f%%")                   #getting the percentage on pie
plt.show()
```
