```python
import os
os.getcwd()

try:
    root_directory=os.getcwd()
    new_folder="bank_graphs_seaborn"
    new_dir=os.path.join(root_directory,new_folder)
    os.makedirs(new_dir)

except Exception as e:
    print(e)
```

```python
for i in cat_columns[1:]:
    plt.figure(figsize=(8,5))
    order_continent=df[i].value_counts().keys()
    sns.countplot(data=df,
                  x=i,
                  order=order_continent)
    plt.title("bar chart")
    plt.savefig(f"{new_dir}\{i} bank_graphs_seaborn.jpg")
    plt.show()
```
