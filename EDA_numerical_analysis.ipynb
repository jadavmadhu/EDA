# Imperical rule

```python
# step:1 = calculate the mean of prevailing_wage
# step:2 = calculate the std of prevailinf_wage
# step:3 = lb=mean-1*std
# step:4 = ub=mean+1*std
# step:5 = con1= data<ub
# step:6 = con2= data>lb
# step:7 = con3= con1 and con2
# step:8 = data[con3]
# step:9 = compare with 68% 25480 count
```

```python
#task1=====get the 68% of data

mean=visa_df["prevailing_wage"].mean()
std=visa_df["prevailing_wage"].std()
lb=mean-1*std
ub=mean+1*std
con=visa_df["prevailing_wage"]<ub
con1=visa_df["prevailing_wage"]>lb
con2=con & con1
visa_df[con2]
```

```python
#task2=====get the 95% of data



mean=visa_df["prevailing_wage"].mean()
std=visa_df["prevailing_wage"].std()
lb=mean-2*std
ub=mean+2*std
con=visa_df["prevailing_wage"]<ub
con1=visa_df["prevailing_wage"]>lb
con2=con & con1
visa_df[con2]
```

```python
#task3=====get the 99.7% of data



mean=visa_df["prevailing_wage"].mean()
std=visa_df["prevailing_wage"].std()
lb=mean-3*std
ub=mean+3*std
con=visa_df["prevailing_wage"]<ub
con1=visa_df["prevailing_wage"]>lb
con2=con & con1
visa_df[con2]
```

```python
wage_data=visa_df["prevailing_wage"]
q1=np.percentile(wage_data,25)
q2=np.percentile(wage_data,50)
q3=np.percentile(wage_data,75)
# Step-2:
IQR= q3-q1
# Step-3:
lb= q1-1.5*IQR
ub=q3+1.5*IQR
#Step-4:
con1=wage_data<lb # T F
con2=wage_data>ub # T F
con3=con1|con2

# Bitwise operators used to compare
# Bits data === 0 and 1 == True false
#Step-3:
outliers_data=wage_data[con3]
outliers_data
```

```python
wage_data=visa_df["prevailing_wage"]
q1=np.percentile(wage_data,25)
q2=np.percentile(wage_data,50)
q3=np.percentile(wage_data,75)
# Step-2:
IQR= q3-q1
# Step-3:
lb= q1-1.5*IQR
ub=q3+1.5*IQR
#Step-4:
con1=wage_data>lb # T F
con2=wage_data<ub # T F
con3=con1&con2

# Bitwise operators used to compare
# Bits data === 0 and 1 == True false
#Step-3:
non_outliers_df=visa_df[con3]
non_outliers_df
```

```python
wage_non_outlier=non_outliers_df["prevailing_wage"]
wage_data=visa_df["prevailing_wage"]
plt.figure(figsize=(14,8))
plt.hist(wage_data)
plt.hist(wage_non_outlier)
plt.show()
```
